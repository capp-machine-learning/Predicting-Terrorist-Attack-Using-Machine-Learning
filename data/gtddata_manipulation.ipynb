```python
import datetime
import pandas as pd
import numpy as np
```

```python
gtd = pd.read_csv("data/globalterrorismdb_0718dist.csv")
gtd.head()
```

```python
gtd2 = gtd[["eventid", "iyear", "imonth", "iday", "latitude", "longitude", "gname"]]
```

```python
filter1 = gtd2["imonth"]!=0
filter2 = gtd2["iday"]!=0
filter3 = gtd2["gname"]!="unknown"
gtd3 = gtd2[filter1][filter2][filter3]
```

```python
gtd3["date"] = gtd3["imonth"].astype(str)+"/"+gtd3["iday"].astype(str)+"/"+gtd3["iyear"].astype(str)
```

```python
date = lambda x: datetime.datetime.strptime(x, '%m/%d/%Y')
gtd3["date"] = gtd3["date"].map(date)
```

```python
lat_lon = gtd3.drop_duplicates(subset=["latitude", "longitude"])[["latitude", "longitude"]]
```

```python
lat_lon["loc_id"] = list(range(len(lat_lon.index)))
```

```python
lat_lon.head()
```

```python
gr= gtd3.drop_duplicates(subset=["gname"])[["gname"]]
```

```python
gr["gr_id"]=list(range(len(gr.index)))
```

```python
gtd4 = pd.merge(gtd3, lat_lon, on = ["latitude", "longitude"], how="inner")
```

```python
gtd5 = pd.merge(gtd4, gr, on = "gname", how = "inner")
```

```python
gtd5= gtd5[["eventid","date","latitude","longitude","loc_id", "gname","gr_id"]]
```
